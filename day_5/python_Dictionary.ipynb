***--> Rules***

`1. Dictionary has no Indexing`

`2. Dictionary is a Mutable Data Type`

`3. Keys -> Immutable & Value -> they can be Mutable`

`4. Key's chould be Unique`

` Mutable --> List/ Dictionary/ Sets`

`Immutable --> Strings/ Tuples / int / float/ Boolean/ Complex`

## **1. --> Create**

`Empty Dictionary`

```python
D = {}
D
```

```python
D1 = {"Name":"Faizan","Age":17}
D1
```

```python
# cannot use 'list' as a dict key
D2 = {[1,2,3]:"Hello"}
D2
```

```python
#  Use Tuple as a Dict key
D2 = {(1,2,3):"Hello"}
D2
```

```python
# Don't Use repeating key's
D3 = {"Name":"Faizan","Name":"Ali"}
D3
```

`2D Dict`

```python
D4 = {"Name":"Faizan","College":"RIC","Marks":{"Eng":50,"Isl":45}}
D4
```

`3D Dict`

```python
D5 = {"DateTime": {"Calendar": {"Date": 16, "Month": 7},"Clock": {"Year": 2026, "Time":5}}}
D5
```

## **4. --> Let's Create a Function**

```python
def is_even(Number):
    if type(Number) == int:
   
        if Number % 2 == 0:
            return "Even"
        else:
            return "Odd"
    else:
        return "Not allowed"

x = is_even(6)
print(x)
```

## **7. --> Nested Function**

```python
def g(x):
    def h(x):
         x = x+1
         print('in h(x): x=',x)
    x = x + 1
    print('in g(x): x=', x)
    h(x)
    return(x)


x = 3
z = g(x)
print('in main program scope: x=',x)
print('in main program scope: z=',z)
```
